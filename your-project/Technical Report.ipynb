## Domain Research
<a id="domain"></a>

*AQUASTAT* is the FAO global information system on water resources and agricultural water management. It collects, analyses and provides free access to over 180 variables and indicators by country from 1960. *AQUASTAT* plays a key role in the monitoring of the Sustainable Development Goal 6 that sets out to "*ensure availability and sustainable management of water and sanitation for all*", and in particular indicators of target 6.4 on water stress and water use efficiency.

## Results
<a id="results"></a>

Different analysis were performed in order to determine and study the evolution of water consumption by humans in the world. First, an initial analysis is made of the evolution of freshwater consumption over the years. Then, the distribution of the renewable water resources per capita is studied by countries. Finally, it's shown the amount and proportion of water withdrawned to use for for agricultural, industrial and municipal activity.

### Global Freshwater Use

The following graph gives a clear image of the the global freshwater consumption evolution through the years. The freshwater use grows steadily at a constant rate till 1950, when it starts increasing drastically. From 1950 to 2015, freshwater use in the world almost quadruples from a value of 1.200.000.000.000 cubic meters to 4.000.000.000.000.

### Renewable Water Resources per Capita

In the geographical study of the distribuition of the renewable water resources per capita, the regions with greater and lower amount of resources can be observed. It is found that regions as Africa, Asia and Europe have the lower water resources per capita available, while South America and the most northern countries of the world have the the highest amounts of resources per inhabitant.

### Water Withdrawal by activity

In the following graph is shown the evolution of water withdrawal by activity. As human activity evolves and changes depending on technological innovations the consumption of water in each sector also varies. It is verified that water withdrawal for agricultural purposes steadily decreases through the years. The industrial sector is the one with the highest water consumption with its peek in 1990 with significant lower consumptions in the latest years. Finally, the municipal sector uses less water than the other two but has experienced a growth in the last years.
